```python
import pandas as pd
import numpy as np
import random
from faker import Faker

fake = Faker()
np.random.seed(42)
random.seed(42)

# --------------------------
# ПАРАМЕТРЫ
# --------------------------
n_orders = 16265
n_customers = 4782
start_date = pd.Timestamp("2023-01-01")
end_date = pd.Timestamp("2024-12-31")

# --------------------------
# АССОРТИМЕНТ
# --------------------------
products = {
    # Laptops
    "MacBook Air 13": (1099, 1199, "Laptop"),
    "MacBook Pro 16": (1999, 2099, "Laptop"),
    "Dell XPS 13": (1199, 1299, "Laptop"),
    "Dell XPS 15": (1499, 1599, "Laptop"),
    "Lenovo ThinkPad X1": (1399, 1499, "Laptop"),
    "HP Spectre x360": (1299, 1399, "Laptop"),
    "Asus ZenBook": (999, 1099, "Laptop"),

    # Phones
    "iPhone 13": (799, 849, "Phone"),
    "iPhone 14": (999, 1049, "Phone"),
    "iPhone 15": (1199, 1249, "Phone"),
    "Samsung Galaxy S21": (749, 799, "Phone"),
    "Samsung Galaxy S23": (949, 999, "Phone"),
    "Xiaomi Redmi Note 12": (349, 399, "Phone"),
    "Xiaomi Mi 11": (499, 549, "Phone"),
    "Google Pixel 7": (699, 749, "Phone"),

    # Tablets
    "iPad 9th Gen": (499, 549, "Tablet"),
    "iPad Pro 12.9": (1099, 1149, "Tablet"),
    "Samsung Galaxy Tab S8": (799, 849, "Tablet"),

    # Accessories
    "Apple AirPods Pro": (249, 259, "Accessory"),
    "Sony WH-1000XM4": (349, 359, "Accessory"),
    "Logitech MX Master Mouse": (99, 109, "Accessory"),
    "Mechanical Keyboard": (149, 159, "Accessory"),
    "USB-C Charger": (29, 34, "Accessory"),
    "Samsung Travel Adapter": (29, 34, "Accessory"),
    "Apple 20W USB-C Power Adapter": (19, 22, "Accessory"),
    "Apple 35W Dual USB-C Adapter": (59, 62, "Accessory"),
    "Apple Mouse": (89, 99, "Accessory"),
    "Apple Pencil": (119, 129, "Accessory"),
    "Screen Protector": (35, 40, "Accessory"),
    "HDMI Cable": (14, 18, "Accessory")
}

payment_methods = ["Card", "Cash", "Online"]
channels = ["Online", "Offline"]

countries = [
    ("USA", 0.25),
    ("China", 0.25),
    ("Russia", 0.15),
    ("Germany", 0.08),
    ("France", 0.07),
    ("Japan", 0.07),
    ("UK", 0.05),
    ("Spain", 0.04),
    ("UAE", 0.02),
    ("South Korea", 0.02)
]
country_names = [c[0] for c in countries]
country_weights = [c[1] for c in countries]

# --------------------------
# ФУНКЦИИ
# --------------------------
def get_quantity(category):
    if category in ["Laptop", "Phone", "Tablet"]:
        return np.random.choice([1, 2], p=[0.92, 0.08])
    else:
        return np.random.choice([1, 2, 3], p=[0.7, 0.2, 0.1])

def dirty_product_name(name):
    r = random.random()
    if r < 0.0005:
        return name.lower()
    elif r < 0.0010:
        return " " + name + " "
    return name

def add_accessories(product, year, base_order):
    accessories = []
    def add_item(name, prob):
        if name not in products:
            return
        if random.random() < prob:
            p2023, p2024, cat = products[name]
            price = p2023 if year == 2023 else p2024
            accessories.append([
                None, base_order['customer_id'], base_order['order_date'],
                dirty_product_name(name), cat,
                1, round(price,2), base_order['payment'], base_order['channel'], base_order['country']
            ])

    # --- логика по категориям ---
    if product.startswith("iPhone 13") or product.startswith("iPhone 14"):
        add_item("Apple 20W USB-C Power Adapter", 0.7)
        add_item("Apple 35W Dual USB-C Adapter", 0.2)
        add_item("Screen Protector", 0.5)
        add_item("Apple AirPods Pro", 0.4)
    elif product.startswith("iPhone 15"):
        add_item("Apple 35W Dual USB-C Adapter", 0.6)
        add_item("Apple 20W USB-C Power Adapter", 0.3)
        add_item("Screen Protector", 0.6)
        add_item("Apple AirPods Pro", 0.5)
    elif "Samsung Galaxy S21" in product or "Samsung Galaxy S23" in product:
        add_item("Samsung Travel Adapter", 0.6)
        add_item("Screen Protector", 0.6)
        add_item("Sony WH-1000XM4", 0.2)
    elif "Xiaomi" in product or "Pixel" in product:
        add_item("USB-C Charger", 0.1)
        add_item("Screen Protector", 0.4)
    elif product.startswith("iPad"):
        add_item("Apple Pencil", 0.5)
        add_item("Apple 20W USB-C Power Adapter", 0.4)
        add_item("Screen Protector", 0.6)
        add_item("Apple AirPods Pro", 0.3)
    elif product.startswith("MacBook"):
        add_item("Apple Mouse", 0.5)
        add_item("Screen Protector", 0.05)
    else:
        add_item("Screen Protector", 0.01)

    return accessories

# --------------------------
# ГЕНЕРАЦИЯ ДАННЫХ
# --------------------------
data = []

# 1) РОЗНИЧНЫЕ ЗАКАЗЫ
for _ in range(n_orders):
    customer_id = random.randint(1, n_customers)
    order_date_py = fake.date_between(start_date=start_date.date(), end_date=end_date.date())
    order_date = pd.Timestamp(order_date_py)

    product, (price_2023, price_2024, category) = random.choice(list(products.items()))
    year = order_date.year
    price = price_2023 if year == 2023 else price_2024
    quantity = get_quantity(category)

    # грязные данные для аксессуаров (редко)
    if category == "Accessory":
        if random.random() < 0.0001:
            price = 0
        if random.random() < 0.0005:
            price = -abs(price)
        if random.random() < 0.0005:
            price = price * 6
            
    product_dirty = dirty_product_name(product)
    payment = random.choice(payment_methods)
    channel = random.choice(channels)
    country = random.choices(country_names, weights=country_weights, k=1)[0]

    base_order = {
        "customer_id": customer_id,
        "order_date": order_date,
        "payment": payment,
        "channel": channel,
        "country": country
    }

    row = [
        None, customer_id, order_date, product_dirty, category,
        quantity, round(price,2), payment, channel, country
    ]
    data.append(row)

    if category in ["Phone", "Tablet", "Laptop"]:
        accessories = add_accessories(product, year, base_order)
        data.extend(accessories)

# 2) ОПТОВЫЕ ЗАКАЗЫ
wholesale_customers = []
wholesale_countries = ["USA"] * 2
for i, country in enumerate(wholesale_countries, start=1):
    wholesale_customers.append({"customer_id": n_customers + i, "country": country})

for wc in wholesale_customers:
    for year in [2023, 2024]:
        for quarter in [1, 2, 3, 4]:
            order_date = pd.Timestamp(f"{year}-{(quarter-1)*3+1:02d}-15")
            for product, (price_2023, price_2024, category) in products.items():
                if category in ["Laptop", "Accessory"]:
                    base_price = price_2023 if year == 2023 else price_2024
                    price = round(base_price * 0.85, 2)
                    quantity = random.choice([20, 25, 30, 35, 40]) if category == "Laptop" else random.choice([30, 40, 50, 60])
                    product_dirty = dirty_product_name(product)
                    data.append([
                        None, wc["customer_id"], order_date, product_dirty, category,
                        quantity, price, "Card", "Offline", wc["country"]
                    ])

# --------------------------
# СОЗДАНИЕ ДАТАФРЕЙМА
# --------------------------
df = pd.DataFrame(data, columns=[
    "order_id", "customer_id", "order_date", "product_name",
    "product_category", "quantity", "price",
    "payment_method", "channel", "country"
])

df['order_date'] = pd.to_datetime(df['order_date'])
df = df.sort_values(by="order_date").reset_index(drop=True)
df["order_id"] = df.index + 1
df["order_date"] = df["order_date"].dt.strftime("%Y-%m-%d")

df.to_csv("rfm_dataset_en.csv", index=False)
print(f"✅ Dataset created with {len(df)} rows")
```

```python
# --- 1) Сводная таблица по продуктам ---
summary = (
    df_clean
    .groupby([df_clean['order_date'].dt.year, 'product_category', 'product_name'])
    .agg(
        retail_price=('price', lambda x: x[df_clean.loc[x.index, 'quantity'] < 20].mean()),
        wholesale_price=('price', lambda x: x[df_clean.loc[x.index, 'quantity'] >= 20].mean())
    )
    .reset_index()
)

# Добавим столбец с разницей в процентах
summary['discount_vs_retail_%'] = (
    (summary['wholesale_price'] / summary['retail_price'] - 1) * 100
).round(2)

print("Сравнение цен по годам, категориям и продуктам:")
display(summary.sort_values(by=['product_category', 'product_name', 'order_date']))

# --- 2) Анализ оптовиков по покупателям ---
wholesale_customers_summary = (
    df_clean[df_clean['quantity'] >= 20]
    .groupby(['customer_id', 'country', df_clean['order_date']])
    .agg(
        total_orders=('order_id', 'nunique'),
        total_quantity=('quantity', 'sum'),
        total_revenue=('price', lambda x: (x * df_clean.loc[x.index, 'quantity']).sum())
    )
    .reset_index()
    .sort_values(by=['order_date'])
)

print("\nОптовые клиенты:")
display(wholesale_customers_summary)
```

```python
# 4.4.6. Динамика количества проданных товаров по месяцам (Phone, Tablet, Laptop)

# Подготовка данных для heatmap по количеству товаров
# Аксессуары не включаем, они могут "перекрывать" картину по основным категориям,
# так как их продается значительно больше по количеству, но меньше по выручке.

# Фильтруем данные только для нужных категорий
filtered_categories = ['Phone', 'Tablet', 'Laptop']
filtered_df = df_new[df_new['product_category'].isin(filtered_categories)]

# Подготовка данных для heatmap по количеству товаров
month_category_quantity = filtered_df.groupby([pd.Grouper(key='order_date', freq='M'), 'product_category'])['quantity'].sum().unstack()

# Форматируем даты в более читаемый вид (год-месяц)
formatted_dates = [date.strftime('%Y-%m') for date in month_category_quantity.index]

plt.figure(figsize=(12, 8))
sns.heatmap(month_category_quantity, 
            cmap='YlGnBu', 
            annot=True, 
            fmt='.0f', 
            linewidths=0.5, 
            cbar_kws={'label': 'Количество товаров'},
            xticklabels=month_category_quantity.columns,
            yticklabels=formatted_dates)

plt.title('Количество проданных товаров по месяцам (Phone, Tablet, Laptop)', fontsize=16, pad=20)
plt.xlabel('Категория товара', fontsize=12)
plt.ylabel('Месяц', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 4.4.7. Расчет AOV (средняя стоимость заказа), распределение AOV и количества заказов по месяцам

# Группируем по месяцам и считаем общую выручку и количество заказов
monthly_stats = df_new.groupby(pd.Grouper(key='order_date', freq='M')).agg(
    total_revenue=('revenue', 'sum'),
    total_orders=('order_id', 'nunique')
).reset_index()

# Рассчитываем AOV (средняя стоимость заказа)
monthly_stats['aov'] = monthly_stats['total_revenue'] / monthly_stats['total_orders']

# Форматируем даты для подписей
monthly_stats['month_formatted'] = monthly_stats['order_date'].dt.strftime('%Y-%m')

print("AOV по месяцам:")
print(monthly_stats[['month_formatted', 'total_orders', 'total_revenue', 'aov']].round(2))
```

```python
fig, ax1 = plt.subplots(figsize=(14, 8))

# Первая ось Y: AOV
color = 'tab:blue'
ax1.set_xlabel('Месяц', fontsize=12)
ax1.set_ylabel('AOV', color=color, fontsize=12)
line = ax1.plot(monthly_stats['month_formatted'], monthly_stats['aov'], 
                color=color, marker='o', linewidth=2.5, label='AOV')
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticklabels(monthly_stats['month_formatted'], rotation=45)

# Добавляем подписи значений только для AOV
for i, (date, aov) in enumerate(zip(monthly_stats['month_formatted'], monthly_stats['aov'])):
    ax1.annotate(f'{aov:.0f}', 
                (i, aov),
                xytext=(0, 5),
                textcoords='offset points',
                ha='left',
                fontsize=13,
                fontweight='bold',
                color='darkblue')

# Вторая ось Y: количество заказов
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Количество заказов', color=color, fontsize=12)
ax2.bar(monthly_stats['month_formatted'], monthly_stats['total_orders'], 
        alpha=0.3, color=color, label='Количество заказов')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('AOV и количество заказов по месяцам', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# Общая статистика по AOV
print("ОБЩАЯ СТАТИСТИКА ПО AOV:")
print("=" * 40)
print(f"Средний AOV за весь период: {monthly_stats['aov'].mean():.2f}")
print(f"Медианный AOV: {monthly_stats['aov'].median():.2f}")
print(f"Минимальный AOV: {monthly_stats['aov'].min():.2f} ({monthly_stats.loc[monthly_stats['aov'].idxmin(), 'month_formatted']})")
print(f"Максимальный AOV: {monthly_stats['aov'].max():.2f} ({monthly_stats.loc[monthly_stats['aov'].idxmax(), 'month_formatted']})")
print(f"Стандартное отклонение: {monthly_stats['aov'].std():.2f}")

# Топ-5 месяцев с самым высоким AOV
print("\nТОП-5 месяцев с самым высоким AOV:")
print("=" * 40)
top_months = monthly_stats.nlargest(5, 'aov')[['month_formatted', 'aov', 'total_orders', 'total_revenue']]
print(top_months.round(2))

# Сравнение роста AOV
first_aov = monthly_stats['aov'].iloc[0]
last_aov = monthly_stats['aov'].iloc[-1]
growth_pct = ((last_aov - first_aov) / first_aov) * 100
print(f"\nРост AOV за период: {growth_pct:.1f}%")
```

   5.1. Создание таблицы с RFM показателями и сегментами
</h3>

```python
# --- 5.1.1. Расчёт R, F, M ---
df_new['order_date'] = pd.to_datetime(df_new['order_date'])
reference_date = df_new['order_date'].max() + timedelta(days=1)

# RFM базовая таблица
rfm = df_new.groupby('customer_id').agg(
    LastPurchase=('order_date', 'max'),
    Frequency=('order_id', 'nunique'),
    Monetary=('revenue', 'sum')
).reset_index()
rfm['Recency'] = (reference_date - rfm['LastPurchase']).dt.days
rfm = rfm[['customer_id', 'Recency', 'Frequency', 'Monetary', 'LastPurchase']]

# --- 5.1.2. Создание сегментов
# Определение бинов ---
R_bins = [-1, 90, 270, np.inf]  # 3 категории

rfm['F_bin'] = pd.qcut(rfm['Frequency'], q=3, labels=False, duplicates='drop') + 1
rfm['M_bin'] = pd.qcut(rfm['Monetary'], q=3, labels=False, duplicates='drop') + 1

# --- Присвоение R, F, M ---
rfm_m = rfm.copy()
rfm_m['R'] = pd.cut(rfm_m['Recency'], bins=R_bins, labels=[1, 2, 3], include_lowest=True).astype(int)
rfm_m['F'] = 4 - rfm_m['F_bin']
rfm_m['M'] = 4 - rfm_m['M_bin']
rfm_m['RFM_Score'] = rfm_m['R'].astype(str) + rfm_m['F'].astype(str) + rfm_m['M'].astype(str)
```

```python
# --- Сегментация ---
def map_segment(code):
    if code == '111':
        return 'Super VIP'
    if code in ('112', '121', '122'):
        return 'Loyal, regular'
    if code in ('211', '212', '221', '222', '231', '232'):
        return 'Potential'
    if code in ('113', '123', '131', '132', '133', '213', '223'):
        return 'Less potential'
    if code in ('312', '321', '322', '331', '332', '313'):
        return 'Sleeping, potential'
    if code == '311':
        return 'Very potential lost'
    return 'Sleeping, low-potential'

rfm_m['Segment'] = rfm_m['RFM_Score'].apply(map_segment)

# --- Статистика по бинам ---
print("\n=== Recency bins summary ===")
print(rfm_m.groupby('R').agg(
    num_customers=('customer_id', 'count'),
    avg_days=('Recency', 'mean'),
    avg_revenue=('Monetary', 'mean')
))

print("\n=== Frequency bins summary ===")
print(rfm_m.groupby('F').agg(
    num_customers=('customer_id', 'count'),
    avg_purchases=('Frequency', 'mean'),
    avg_revenue=('Monetary', 'mean')
))

print("\n=== Monetary bins summary ===")
print(rfm_m.groupby('M').agg(
    num_customers=('customer_id', 'count'),
    avg_revenue=('Monetary', 'mean'),
    median_revenue=('Monetary', 'median')
))

# --- Сводка ---
summary = rfm_m.groupby(['Segment', 'RFM_Score']).agg(
    num_customers=('customer_id', 'count'),
    total_revenue=('Monetary', 'sum'),
    avg_revenue=('Monetary', 'mean'),
    median_revenue=('Monetary', 'median')
).reset_index()

total_customers = rfm_m['customer_id'].nunique()
total_revenue = rfm_m['Monetary'].sum()
summary['customer_share_%'] = (summary['num_customers'] / total_customers * 100).round(1)
summary['revenue_share_%'] = (summary['total_revenue'] / total_revenue * 100).round(1)
summary = summary.sort_values('revenue_share_%', ascending=False)
```

# **Финальный вывод по проекту**

На основе выполненного анализа с использованием **сгенерированных данных**, а также в результате очистки данных, проведения EDA и RFM-анализа были получены следующие ключевые выводы и рекомендации.

---

## **Ключевые выводы**

### 1️⃣ Здоровый баланс бизнес-модели
Бизнес успешно сочетает **volume-стратегию** (массовые продажи аксессуаров) с **value-стратегией** (премиум-техника), обеспечивая стабильный cash flow и высокую маржинальность.

### 2️⃣ Географическая стабильность
Топ-3 рынка (**США, Китай, Россия**) демонстрируют идентичное поведение клиентов по всем метрикам, что подтверждает универсальность товарного предложения и эффективность глобальной стратегии.

### 3️⃣ Четкая сегментация клиентской базы
RFM-анализ выявил **7 сегментов** с разным потенциалом и ценностью:

**🔍 Критические находки из RFM-анализа**

**Высокоприоритетные сегменты (ядро бизнеса):**
- **Super VIP** (535 клиентов, 11.5% базы) → 2.84M USD выручки (22.6%), avg чек 5,301 USD  
- **Potential** (1,192 клиентов, 25.7% базы) → 3.68M USD выручки (29.3%), avg чек 3,089 USD  

**Средний приоритет:**
- **Very potential lost** (85 клиентов, 1.8%) → 0.40M USD (3.2%), avg чек 4,713 USD  
- **Loyal, regular** (678 клиентов, 14.6%) → 2.05M USD (16.4%), avg чек 3,021 USD  

**Низкоценные сегменты:**
- **Sleeping, potential** (435 клиентов, 9.4%) → 1.06M USD (8.5%), avg чек 2,442 USD  
- **Sleeping, low-potential** + **Less potential** (1,665 клиентов, 35.9%) → 1.44M USD (11.5%), avg чек 778–1,017 USD  

---

## **🚀 Стратегические рекомендации**

### Приоритет 1: Реактивация "спящих" клиентов
- **Sleeping, potential** → персональные офферы, email-рассылки, VIP-программы  
- **Very potential lost** → эксклюзивные скидки, персональные менеджеры  
- **KPI:** вернуть 20% этих клиентов → +8–10% к общей выручке  

### Приоритет 2: Удержание VIP и перспективных клиентов
- **Super VIP + Potential** → персональные условия, ранний доступ к новинкам  
- **Кросс-селл**: аксессуары и дополнения к основной технике  
- **Программа лояльности** для Loyal, regular → стимулирование апсейла и повторных покупок  

### Приоритет 3: Оптимизация товарного ассортимента
- **Фокус на премиум-технике**: MacBook Pro, iPhone 15  
- **Создание связок**: телефон/ноутбук/планшет + пленка/адаптер
- **Сезонное планирование запасов**: ноутбуки к ноябрю, телефоны к августу  

### Приоритет 4: Развитие онлайн-канала
- **Улучшение рекомендательной системы** на основе товарных связок  
- **Упрощение процесса покупки** аксессуаров при основном заказе  
- **Внедрение видео-консультаций** для премиум-техники  

---

## **📈 Метрики успеха**

**Короткий срок (3 месяца):**  
- Рост доли онлайн-продаж на 15%  
- Реактивация 20% "спящих" ценных клиентов  

**Средний срок (6 месяцев):**  
- Рост AOV на 10% через кросс-селл  
- Увеличение повторных покупок на 25%  

**Долгий срок (12 месяцев):**  
- Рост общей выручки на 20%  
- Увеличение LTV на 30%  

---

## **💡 Итог**
Бизнес демонстрирует **здоровую модель с четкой сегментацией клиентов и товаров**.  
Главная точка роста — **работа с существующими клиентами**: реактивация "спящих", персонализация офферов и увеличение среднего чека через кросс-селл.  

**Фокус на ближайшие 6 месяцев:** реактивация ценных "спящих" клиентов и увеличение среднего чека через кросс-селл аксессуаров к основной технике.  

Проект подтверждает, что компания имеет все компоненты для ускоренного роста — усилия следует направить на ключевые сегменты и товары с высокой маржинальностью.
