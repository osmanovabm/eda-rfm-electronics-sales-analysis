```python
import pandas as pd
import numpy as np
import random
from faker import Faker

fake = Faker()
np.random.seed(42)
random.seed(42)

# --------------------------
# PARAMETERS
# --------------------------
n_orders = 16265
n_customers = 4782
start_date = pd.Timestamp("2023-01-01")
end_date = pd.Timestamp("2024-12-31")

# --------------------------
# PRODUCT ASSORTMENT
# --------------------------
products = {
    # Laptops
    "MacBook Air 13": (1099, 1199, "Laptop"),
    "MacBook Pro 16": (1999, 2099, "Laptop"),
    "Dell XPS 13": (1199, 1299, "Laptop"),
    "Dell XPS 15": (1499, 1599, "Laptop"),
    "Lenovo ThinkPad X1": (1399, 1499, "Laptop"),
    "HP Spectre x360": (1299, 1399, "Laptop"),
    "Asus ZenBook": (999, 1099, "Laptop"),

    # Phones
    "iPhone 13": (799, 849, "Phone"),
    "iPhone 14": (999, 1049, "Phone"),
    "iPhone 15": (1199, 1249, "Phone"),
    "Samsung Galaxy S21": (749, 799, "Phone"),
    "Samsung Galaxy S23": (949, 999, "Phone"),
    "Xiaomi Redmi Note 12": (349, 399, "Phone"),
    "Xiaomi Mi 11": (499, 549, "Phone"),
    "Google Pixel 7": (699, 749, "Phone"),

    # Tablets
    "iPad 9th Gen": (499, 549, "Tablet"),
    "iPad Pro 12.9": (1099, 1149, "Tablet"),
    "Samsung Galaxy Tab S8": (799, 849, "Tablet"),

    # Accessories
    "Apple AirPods Pro": (249, 259, "Accessory"),
    "Sony WH-1000XM4": (349, 359, "Accessory"),
    "Logitech MX Master Mouse": (99, 109, "Accessory"),
    "Mechanical Keyboard": (149, 159, "Accessory"),
    "USB-C Charger": (29, 34, "Accessory"),
    "Samsung Travel Adapter": (29, 34, "Accessory"),
    "Apple 20W USB-C Power Adapter": (19, 22, "Accessory"),
    "Apple 35W Dual USB-C Adapter": (59, 62, "Accessory"),
    "Apple Mouse": (89, 99, "Accessory"),
    "Apple Pencil": (119, 129, "Accessory"),
    "Screen Protector": (35, 40, "Accessory"),
    "HDMI Cable": (14, 18, "Accessory")
}

payment_methods = ["Card", "Cash", "Online"]
channels = ["Online", "Offline"]

countries = [
    ("USA", 0.25),
    ("China", 0.25),
    ("Russia", 0.15),
    ("Germany", 0.08),
    ("France", 0.07),
    ("Japan", 0.07),
    ("UK", 0.05),
    ("Spain", 0.04),
    ("UAE", 0.02),
    ("South Korea", 0.02)
]
country_names = [c[0] for c in countries]
country_weights = [c[1] for c in countries]

# --------------------------
# FUNCTIONS
# --------------------------
def get_quantity(category):
    # Define typical quantity for each category
    if category in ["Laptop", "Phone", "Tablet"]:
        return np.random.choice([1, 2], p=[0.92, 0.08])
    else:
        return np.random.choice([1, 2, 3], p=[0.7, 0.2, 0.1])

def dirty_product_name(name):
    # Introduce small random data inconsistencies
    r = random.random()
    if r < 0.0005:
        return name.lower()
    elif r < 0.0010:
        return " " + name + " "
    return name

def add_accessories(product, year, base_order):
    # Generate accessory items based on main product
    accessories = []
    def add_item(name, prob):
        if name not in products:
            return
        if random.random() < prob:
            p2023, p2024, cat = products[name]
            price = p2023 if year == 2023 else p2024
            accessories.append([
                None, base_order['customer_id'], base_order['order_date'],
                dirty_product_name(name), cat,
                1, round(price,2), base_order['payment'], base_order['channel'], base_order['country']
            ])

    # --- Category-based accessory logic ---
    if product.startswith("iPhone 13") or product.startswith("iPhone 14"):
        add_item("Apple 20W USB-C Power Adapter", 0.7)
        add_item("Apple 35W Dual USB-C Adapter", 0.2)
        add_item("Screen Protector", 0.5)
        add_item("Apple AirPods Pro", 0.4)
    elif product.startswith("iPhone 15"):
        add_item("Apple 35W Dual USB-C Adapter", 0.6)
        add_item("Apple 20W USB-C Power Adapter", 0.3)
        add_item("Screen Protector", 0.6)
        add_item("Apple AirPods Pro", 0.5)
    elif "Samsung Galaxy S21" in product or "Samsung Galaxy S23" in product:
        add_item("Samsung Travel Adapter", 0.6)
        add_item("Screen Protector", 0.6)
        add_item("Sony WH-1000XM4", 0.2)
    elif "Xiaomi" in product or "Pixel" in product:
        add_item("USB-C Charger", 0.1)
        add_item("Screen Protector", 0.4)
    elif product.startswith("iPad"):
        add_item("Apple Pencil", 0.5)
        add_item("Apple 20W USB-C Power Adapter", 0.4)
        add_item("Screen Protector", 0.6)
        add_item("Apple AirPods Pro", 0.3)
    elif product.startswith("MacBook"):
        add_item("Apple Mouse", 0.5)
        add_item("Screen Protector", 0.05)
    else:
        add_item("Screen Protector", 0.01)

    return accessories

# --------------------------
# DATA GENERATION
# --------------------------
data = []

# 1) RETAIL ORDERS
for _ in range(n_orders):
    customer_id = random.randint(1, n_customers)
    order_date_py = fake.date_between(start_date=start_date.date(), end_date=end_date.date())
    order_date = pd.Timestamp(order_date_py)

    product, (price_2023, price_2024, category) = random.choice(list(products.items()))
    year = order_date.year
    price = price_2023 if year == 2023 else price_2024
    quantity = get_quantity(category)

    # Rare dirty data for accessories
    if category == "Accessory":
        if random.random() < 0.0001:
            price = 0
        if random.random() < 0.0005:
            price = -abs(price)
        if random.random() < 0.0005:
            price = price * 6
            
    product_dirty = dirty_product_name(product)
    payment = random.choice(payment_methods)
    channel = random.choice(channels)
    country = random.choices(country_names, weights=country_weights, k=1)[0]

    base_order = {
        "customer_id": customer_id,
        "order_date": order_date,
        "payment": payment,
        "channel": channel,
        "country": country
    }

    row = [
        None, customer_id, order_date, product_dirty, category,
        quantity, round(price,2), payment, channel, country
    ]
    data.append(row)

    if category in ["Phone", "Tablet", "Laptop"]:
        accessories = add_accessories(product, year, base_order)
        data.extend(accessories)

# 2) WHOLESALE ORDERS
wholesale_customers = []
wholesale_countries = ["USA"] * 2
for i, country in enumerate(wholesale_countries, start=1):
    wholesale_customers.append({"customer_id": n_customers + i, "country": country})

for wc in wholesale_customers:
    for year in [2023, 2024]:
        for quarter in [1, 2, 3, 4]:
            order_date = pd.Timestamp(f"{year}-{(quarter-1)*3+1:02d}-15")
            for product, (price_2023, price_2024, category) in products.items():
                if category in ["Laptop", "Accessory"]:
                    base_price = price_2023 if year == 2023 else price_2024
                    price = round(base_price * 0.85, 2)
                    quantity = random.choice([20, 25, 30, 35, 40]) if category == "Laptop" else random.choice([30, 40, 50, 60])
                    product_dirty = dirty_product_name(product)
                    data.append([
                        None, wc["customer_id"], order_date, product_dirty, category,
                        quantity, price, "Card", "Offline", wc["country"]
                    ])

# --------------------------
# CREATE DATAFRAME
# --------------------------
df = pd.DataFrame(data, columns=[
    "order_id", "customer_id", "order_date", "product_name",
    "product_category", "quantity", "price",
    "payment_method", "channel", "country"
])

df['order_date'] = pd.to_datetime(df['order_date'])
df = df.sort_values(by="order_date").reset_index(drop=True)
df["order_id"] = df.index + 1
df["order_date"] = df["order_date"].dt.strftime("%Y-%m-%d")

df.to_csv("rfm_dataset_en.csv", index=False)
print(f"✅ Dataset created with {len(df)} rows")
```

```python
# Check if the price varies depending on the quantity of purchased items
print(df_clean['quantity'].sort_values().unique())

orders_by_quantity = df_clean.groupby('quantity')['order_id'].count().reset_index()
orders_by_quantity = orders_by_quantity.rename(columns={'order_id': 'order_count'})
orders_by_quantity = orders_by_quantity.sort_values(by='quantity')
print(orders_by_quantity)
```

```python
# --- 1) Pivot table by products ---
summary = (
    df_clean
    .groupby([df_clean['order_date'].dt.year, 'product_category', 'product_name'])
    .agg(
        retail_price=('price', lambda x: x[df_clean.loc[x.index, 'quantity'] < 20].mean()),
        wholesale_price=('price', lambda x: x[df_clean.loc[x.index, 'quantity'] >= 20].mean())
    )
    .reset_index()
)

# Add a column with the percentage difference
summary['discount_vs_retail_%'] = (
    (summary['wholesale_price'] / summary['retail_price'] - 1) * 100
).round(2)

print("Price comparison by year, category, and product:")
display(summary.sort_values(by=['product_category', 'product_name', 'order_date']))

# --- 2) Analysis of wholesale customers ---
wholesale_customers_summary = (
    df_clean[df_clean['quantity'] >= 20]
    .groupby(['customer_id', 'country', df_clean['order_date']])
    .agg(
        total_orders=('order_id', 'nunique'),
        total_quantity=('quantity', 'sum'),
        total_revenue=('price', lambda x: (x * df_clean.loc[x.index, 'quantity']).sum())
    )
    .reset_index()
    .sort_values(by=['order_date'])
)

print("\nWholesale customers:")
display(wholesale_customers_summary)
```

*Finding:*<br>
Records with anomalous prices were found: zero and negative values (for example, -14.0 for 'HDMI Cable', 0.0 for 'Apple Mouse'). It was also identified that customers with `customer_id` 4783 and 4784 made large wholesale orders 4 times a year (quantity >= 20), for which a 15% wholesale discount applied.

<br>
*Solution:*<br>
A strategic decision was made:<br>
- Remove wholesale orders (quantity >= 20) from the dataset, since the goal of this EDA and RFM analysis is retail customer segmentation. Their purchase patterns and motivations are fundamentally different.<br>
- Correct anomalous prices for retail purchases. Based on the analysis of correct prices for the corresponding year, erroneous values were replaced with the correct ones (for example, -14.0 -> 14.0 for HDMI Cable in 2023).<br>

After cleaning, the dataset contains only retail purchases (quantity < 20) with correct prices.

```python
# 4.4.3. Analysis of Top Products by Category
categories = df_new['product_category'].unique()

# Create subplot for each category
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, category in enumerate(categories):
    # Filter data by category
    category_df = df_new[df_new['product_category'] == category]
    
    # Group by products and calculate metrics
    product_stats = category_df.groupby('product_name').agg({
        'quantity': 'sum',
        'revenue': 'sum',
        'order_id': 'count'
    }).nlargest(8, 'quantity')  # Top-8 products by quantity
    
    # Create bar chart
    bars = axes[i].bar(product_stats.index, product_stats['quantity'], 
                       color='lightblue', edgecolor='navy', alpha=0.7)
    
    axes[i].set_title(f'Top Products in Category: {category}', fontsize=14, pad=10)
    axes[i].set_ylabel('Number of Sales', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)
    
    # Add labels with quantity and revenue
    for j, (idx, row) in enumerate(product_stats.iterrows()):
        axes[i].text(j, row['quantity'] + max(product_stats['quantity']) * 0.01, 
                     f'{row["quantity"]:.0f} pcs\n${row["revenue"] / 1000 :.0f}k', 
                     ha='center', va='bottom', fontsize=11,
                     bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

plt.tight_layout()
plt.show()
```

```python
# -----------------------
# 4.4.4. Relationship between Main Products and Accessories
# -----------------------
main_categories = ['Laptop', 'Phone', 'Tablet']
accessory_df = df_new[df_new['product_category'] == 'Accessory']

merged = df_new[df_new['product_category'].isin(main_categories)].merge(
    accessory_df, on='customer_id', suffixes=('_main','_accessory')
)

cross_tab = pd.crosstab(merged['product_category_main'], merged['product_name_accessory'])
plt.figure(figsize=(12,6))
sns.heatmap(cross_tab, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Accessory Cross-Sales by Main Product Categories')
plt.show()
```

```python
# 4.4.7. AOV Calculation (Average Order Value), AOV Distribution and Monthly Order Counts

# Group by month and calculate total revenue and total orders
monthly_stats = df_new.groupby(pd.Grouper(key='order_date', freq='M')).agg(
    total_revenue=('revenue', 'sum'),
    total_orders=('order_id', 'nunique')
).reset_index()

# Calculate AOV (Average Order Value)
monthly_stats['aov'] = monthly_stats['total_revenue'] / monthly_stats['total_orders']

# Format dates for labels
monthly_stats['month_formatted'] = monthly_stats['order_date'].dt.strftime('%Y-%m')

print("Monthly AOV:")
print(monthly_stats[['month_formatted', 'total_orders', 'total_revenue', 'aov']].round(2))
```

   5.1. Creating the RFM Table with Segments

```python
# --- 5.1.1. Calculating R, F, M ---
df_new['order_date'] = pd.to_datetime(df_new['order_date'])
reference_date = df_new['order_date'].max() + timedelta(days=1)

# Base RFM table
rfm = df_new.groupby('customer_id').agg(
    LastPurchase=('order_date', 'max'),
    Frequency=('order_id', 'nunique'),
    Monetary=('revenue', 'sum')
).reset_index()
rfm['Recency'] = (reference_date - rfm['LastPurchase']).dt.days
rfm = rfm[['customer_id', 'Recency', 'Frequency', 'Monetary', 'LastPurchase']]

# --- 5.1.2. Creating Segments ---
# Define bins ---
R_bins = [-1, 90, 270, np.inf]  # 3 categories

rfm['F_bin'] = pd.qcut(rfm['Frequency'], q=3, labels=False, duplicates='drop') + 1
rfm['M_bin'] = pd.qcut(rfm['Monetary'], q=3, labels=False, duplicates='drop') + 1

# --- Assigning R, F, M ---
rfm_m = rfm.copy()
rfm_m['R'] = pd.cut(rfm_m['Recency'], bins=R_bins, labels=[1, 2, 3], include_lowest=True).astype(int)
rfm_m['F'] = 4 - rfm_m['F_bin']
rfm_m['M'] = 4 - rfm_m['M_bin']
rfm_m['RFM_Score'] = rfm_m['R'].astype(str) + rfm_m['F'].astype(str) + rfm_m['M'].astype(str)
```

```python
# --- Segmentation ---
def map_segment(code):
    if code == '111':
        return 'Super VIP'
    if code in ('112', '121', '122'):
        return 'Loyal, regular'
    if code in ('211', '212', '221', '222', '231', '232'):
        return 'Potential'
    if code in ('113', '123', '131', '132', '133', '213', '223'):
        return 'Less potential'
    if code in ('312', '321', '322', '331', '332', '313'):
        return 'Sleeping, potential'
    if code == '311':
        return 'Very potential lost'
    return 'Sleeping, low-potential'

rfm_m['Segment'] = rfm_m['RFM_Score'].apply(map_segment)

# --- Bin statistics ---
print("\n=== Recency bins summary ===")
print(rfm_m.groupby('R').agg(
    num_customers=('customer_id', 'count'),
    avg_days=('Recency', 'mean'),
    avg_revenue=('Monetary', 'mean')
))

print("\n=== Frequency bins summary ===")
print(rfm_m.groupby('F').agg(
    num_customers=('customer_id', 'count'),
    avg_purchases=('Frequency', 'mean'),
    avg_revenue=('Monetary', 'mean')
))

print("\n=== Monetary bins summary ===")
print(rfm_m.groupby('M').agg(
    num_customers=('customer_id', 'count'),
    avg_revenue=('Monetary', 'mean'),
    median_revenue=('Monetary', 'median')
))

# --- Summary table ---
summary = rfm_m.groupby(['Segment', 'RFM_Score']).agg(
    num_customers=('customer_id', 'count'),
    total_revenue=('Monetary', 'sum'),
    avg_revenue=('Monetary', 'mean'),
    median_revenue=('Monetary', 'median')
).reset_index()

total_customers = rfm_m['customer_id'].nunique()
total_revenue = rfm_m['Monetary'].sum()
summary['customer_share_%'] = (summary['num_customers'] / total_customers * 100).round(1)
summary['revenue_share_%'] = (summary['total_revenue'] / total_revenue * 100).round(1)
summary = summary.sort_values('revenue_share_%', ascending=False)
```
